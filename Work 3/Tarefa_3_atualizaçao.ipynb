```python
#Inicializando os nossos parâmetros de pesos e vieses

def initialize_weights():
    
    #Os pesos são inicializados com o random randn e os vieses com np zeros para anular qualquer enviesamento inicial
    W1 = np.random.randn(49, 10)
    b1 = np.zeros((1, 10))
    
    W2 = np.random.randn(49, 10)
    b2 = np.zeros((1, 49))
    
    W3 = np.random.randn(49, 6)
    b3 = np.zeros((1, 6))

    return W1, W2, W3, b1, b2, b3
```

```python
# Criando a Foward Propagation usando a função de ativação sigmoid
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))
# Derivada da função de ativação sigmoid
def sigmoid_derivative(Z):
    return sigmoid(Z) * (1 - sigmoid(Z))

def forward_propagation(X, W1, W2, W3, b1, b2, b3):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)
    Z2 = np.dot(A1, W2.T) + b2
    A2 = sigmoid(Z2)
    Z3 = np.dot(A2, W3) + b3
    output_layer = sigmoid(Z3)

    return Z1, Z2, Z3, A1, A2, output_layer

def back_propagation(X, y, output_layer, W3, W2, W1, A2, A1, b3, b2, b1, learning_rate):
    output_error = y - output_layer
    output_delta = output_error * sigmoid_derivative(output_layer)

    hidden_error2 = output_delta.dot(W3.T)
    hidden_delta2 = hidden_error2 * sigmoid_derivative(A2)

    hidden_error1 = hidden_delta2.dot(W2.T)
    hidden_delta1 = hidden_error1 * sigmoid_derivative(A1)

    W3 += A2.T.dot(output_delta) * learning_rate
    W2 += A1.T.dot(hidden_delta2) * learning_rate
    W1 += X.T.dot(hidden_delta1) * learning_rate
    b3 += np.sum(output_delta, axis=0) * learning_rate
    b2 += np.sum(A2, axis=0) * learning_rate
    b1 += np.sum(A1, axis=0) * learning_rate

    return W1, W2, W3, b1, b2, b3
```
